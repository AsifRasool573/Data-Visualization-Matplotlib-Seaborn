### Functional Plot

```python
#basic line plot, can be used anywhere for initial or bird eye view analysis of features.

x_1 = np.linspace(0,5,10)
y_1 = x_1 ** 2

# print(x_1)
# print(y_1)

plt.plot(x_1,y_1,  marker='.')
plt.title('Days Squard Chart')
plt.xlabel('Days')
plt.ylabel('Days Squared')
```

### Printing Multiple Plots

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_1, 'r')

plt.subplot(1,2,2)
plt.plot(y_1,x_1,'b')
plt.show()
```

### Using Figure Objects


```python
#plotting with axes and fig

fig_1 = plt.figure(figsize=(5,4), dpi=100)       #figsize determins the width and height of the image.
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])      #( [left, bottom, width, height] ) for axis
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')
axes_1.plot(x_1,y_1,label='x/x sq.', marker='.')
axes_1.plot(y_1,x_1,label='x sq./x', marker='.')
axes_1.legend(loc='best')


# #plot within the bigger plot
axes_2= fig_1.add_axes([0.45,0.45,0.4,0.3])      #if you want to take it out side of the first plot. then increase the left value
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(x_1,y_1, marker='o')

axes_2.text(0,20,'Whatever Text')
```

```python
## Stacked bar plot
t_type = ['Kinder', 'Elementry', 'Secondary', 'Spec. Edu.']
m_teach=np.array([22,30,34,14])
f_teach = np.array([60,50,46,76])
d_teach = np.array([18,20,20,10])

ind = [x for x, _ in enumerate(t_type)]    # list comprehension and enumerate to fetch the index of t_type
# ind

plt.bar(ind, m_teach, width=0.45, label = 'Male', bottom=f_teach, color='r')
plt.bar(ind, f_teach, width = 0.45, label = 'Female', bottom = d_teach, color ='g')
plt.bar(ind, d_teach, width = 0.45, label = 'Diverse', color ='b')
plt.legend(loc='lower right')
```

### Pie Charts

```python
# Pie charts are best to use when you are trying to compare parts of a whole. They do not show changes over time.
# When you have small data.

#Basic
exp_vals = [1400,600,300,410,250]
exp_labels = ['Home Rent','Food','Phone/Internet Bill', 'Car', 'Others']
colors = ['red','LightCoral','BlueViolet','yellowgreen','black']
explode=[0.1,0.1,0,0,0.1]
# print(max(exp_vals))

plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%', shadow=True, explode=explode, colors=colors)
# , shadow=True

plt.legend(exp_labels, loc ='upper right', bbox_to_anchor=(1,0,1.1,1))
plt.show()
# Radius is the size of the pie chart
# autopct is to show the percentage and its roundoff number
# explode is for show how much you want the piece of pie to go out.
```

### Heatmaps

```python
# Color coded representation of data from 2-D list.

# Heatmaps are used to show relationships between two variables, one plotted on each axis

symptoms = ["Coronavirus","Infuenza", "Pneunomia","Dyspnea"]
dates = ["Jun-28", "Jul-05","Jul-12","Jul-19","Jul-26","Aug-02","Aug-09","Aug-16","Aug-21"]

symp_per = np.array([[5.2,5.5,5.7,5.6,5.3,5.1,5.0,4.9,5.3],
                    [3.5,4.0,4.3,3.9,3.5,3.2,2.7,2.2,2.0],
                    [1.8,2.2,2.3,2.2,2.1,1.9,1.7,1.4,1.3],
                    [1.0,1.1,1.1,1.0,0.9,0.8,0.8,0.8,0.7]])


fig_10 = plt.figure(figsize=(5,4), dpi=100)       #figsize determins the width and height of the image.
axes_10 = fig_10.add_axes([0.1,0.1,0.9,0.9])

# fig_10, axes_10 = plt.subplots()

# imshow: Display data as an image, i.e., on a 2D regular raster.
im = axes_10.imshow(symp_per, cmap='twilight')    #imshow is going to define the data which plot out all of this information

# cmap: twilight, Wivistia, summer, winter, plasma, and many more can be found on the following link
# https://matplotlib.org/stable/tutorials/colors/colormaps.html


#now adding numbers and label information

axes_10.set_xticks(np.arange(len(dates)))
axes_10.set_yticks(np.arange(len(symptoms)))

axes_10.set_xticklabels(dates)
axes_10.set_yticklabels(symptoms)


# not to make the dates run into each others
# Setp(): Set one or more properties on an Artist, or list allowed values.
plt.setp(axes_10.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# rotation_mode="anchor" ,"default": aligns the unrotated text and then rotates the text around the point of alignment. default
# is like none

#now adding data over it to analyse it.
#looping over data to make annotation of the data.

for i in range(len(symptoms)):
    for j in range(len(dates)):
        text = axes_10.text(j,i,symp_per[i,j],
                           ha="center", va="center", color ="k")

# Axes.text(x, y, s, fontdict=None, **kwargs) **kwargs means other arguments. (a lot of arguments)
# x,y: The position to place the text. By default, this is in data coordinates.
# s: string(): the text.

# horizontalalignment or ha : {'center', 'right', 'left'}
# verticalalignment or va: {'center', 'top', 'bottom', 'baseline', 'center_baseline'}

plt.show()
```
